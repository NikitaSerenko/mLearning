```python
import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''
from scipy import diag

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n,l) = F.shape
    print n, l
    Sigma = 1.0 / l * F.dot(np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    V, d, U = np.linalg.svd(Sigma)
    
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot for i in range(n)])
    
    m = sum(var_i < varRetained)
    accuracy = (np.sum(d[:m]) / dTot * 100.0)
    print '%.2f %% variance retained in %d dimensions' % (accuracy, m)
    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    U_reduced = U[:, :m] 
    
    G = U_reduced.T.dot(F)
    return G, U_reduced
```

```python
img = plt.imread('dreamstime_xs_2293918.jpg')
plt.figure(figsize=(10, 10))
plt.imshow(img) 
```

```python
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering
```

```python
from skimage.measure import compare_ssim
```

```python
def convert(img, cluster_type):
    img_data = np.copy(img).reshape(-1, 3)
    clusters = cluster_type.fit_predict(img_data)
    for i, x in enumerate(np.unique(clusters)):
        img_data[clusters == x] = np.mean(img_data[clusters == x], axis=0)
    converted_img = img_data.reshape(img.shape)
    plt.figure(figsize=(10, 10))
    plt.imshow(converted_img)
    plt.show()
    return compare_ssim(img, converted_img, multichannel=True)
```

```python
k_list = [2, 3, 10, 20]
for k in k_list:
    print convert(img, KMeans(n_clusters=k))
```
